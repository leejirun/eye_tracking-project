```python
#open cv설치
#open cv란? 다양한 영상/동영상 처리에 사용할 수 있는 오픈소스 라이브러리
```

```python
import cv2
import numpy as np #행렬연산
import matplotlib.pyplot as plt #그래프 
```

```python
#이미지 로드
#rgb로 한 픽셀[255,255,255]
#cv2.imread('./pikachu.png') 이미지를 읽기 위해서는 imread() 함수

#이미지 속성
img = cv2.imread('./image/test.jpg') 
print(img.shape)

#opnecv의 imread를 통해서 이미지를 읽으면 rgb값을 거꾸로 bgr로 바꿔서 rgb로 바꿔야 제대로 색이 출력
#이미지 출력 matplotlib 을 통해서 이미지 출력해서 이미지 확인하기
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)# cv2.cvtcolor(원본 이미지, 색상 변환 코드)를 이용하여 이미지의 색상 공간을 변경
plt.imshow(rgb_img)
```

```python
height, width, channel = img.shape #해당 이미지의 높이, 너비, 채널의 값을 확인
print(img.shape)
```

```python
print('높이', height, '넒이', width, '채널(색상차원)', channel) #channel이 3일경우 다색(컬러)이미지 1인경우 흑백사진, (0~255,0~255,0~255) 
```

```python
#해당 영역만 나오도록 자르기
plt.imshow((rgb_img[310:570, 250:470])) 
#(y축, x축)이미지에서  y축 310~570, x축 250~470, 시작점(250,310) : 끝점(470.570)
#너비 : 220, 높이:  260
```

```python
#opencv 사각형 그리기
x,y,w,h = 250,310,220,260
#파라미터 순서 -> 그릴 이미지, 시작점, 끝점, 색상, 선두께
rect_img = cv2.rectangle(rgb_img,(x,y),(x+w,y+h), (255,0,0),3)
plt.imshow(rect_img)
```

```python
#그림에 글지 넣기
# 파라미터 순서 -> 글자를 넣을 이미지, 넣을 글자, 시작점, 글자폰트, 글자크기, 글자색상, 글자 두께, 선종류
text_img = cv2.putText(rect_img, 'mysein', (x,y-50), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,0,255), 3, cv2.LINE_AA)
plt.imshow(text_img)
```
